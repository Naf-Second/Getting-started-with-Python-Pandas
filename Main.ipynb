```python
def A1_Task4_2020_2_60_175(column_name1):
  df.boxplot(column = column_name1)

  sorted_values = np.sort(df[column_name1])
  Q1 = np.percentile(sorted_values, 25)
  median = np.percentile(sorted_values, 50)
  Q3 = np.percentile(sorted_values, 75)


  IQR = Q3 - Q1
  lb = Q1 - 1.5 * IQR
  ub = Q3 + 1.5 * IQR

  outliers = []
  for value in df[column_name1]:
      if value < lb or value > ub:
          outliers.append(value)

  print(f"Q1: {Q1}")
  print(f"Median (Q2): {median}")
  print(f"Q3: {Q3}")
  print(f"IQR: {IQR}")
  print(f'Number of Outliers {len(outliers)}')
```

```python
print("Enter column name: ")
col = input()
A1_Task4_2020_2_60_175(col)
```

```python
def A1_Task5_2020_2_60_175(column_name1):
  df.hist(column = column_name1)
  print(df[column_name1].kurt())
  print(df[column_name1].skew())
  fx = df[column_name1].kurt()
  if fx>=1 :
    print("Leptokurtic")
  elif fx>0 and fx<1 :
    print('Mesokurtic')
  else :
    print("Platykurtic")
```

```python
def A1_Task10_2020_2_60_175():
    maxcor = 0
    col1 = ''
    col2 = ''
    for i in range(df.shape[1]):
       for j in range(i + 1, df.shape[1]):
            correlation = df.iloc[:, i].corr(df.iloc[:, j])
            if(correlation>maxcor):
              maxcor = correlation
              col1 = df.columns[i]
              col2 = df.columns[j]
    print(f'Highest positive correlation {maxcor} between {col1} and {col2} ')
    return col1,col2
```
